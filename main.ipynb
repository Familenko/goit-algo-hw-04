```python
import random
import timeit
import pandas as pd
```

```python
def insertion_sort(lst):
    for i in range(1, len(lst)):
        key = lst[i]
        j = i-1
        while j >=0 and key < lst[j] :
                lst[j+1] = lst[j]
                j -= 1
        lst[j+1] = key 
    return lst
```

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    return merge(merge_sort(left_half), merge_sort(right_half))

def merge(left, right):
    merged = []
    left_index = 0
    right_index = 0

    while left_index < len(left) and right_index < len(right):
        if left[left_index] <= right[right_index]:
            merged.append(left[left_index])
            left_index += 1
        else:
            merged.append(right[right_index])
            right_index += 1

    while left_index < len(left):
        merged.append(left[left_index])
        left_index += 1

    while right_index < len(right):
        merged.append(right[right_index])
        right_index += 1

    return merged
```

```python
def timsort(arr):
    min_run = 32
    n = len(arr)

    for i in range(0, n, min_run):
        insertion_sort(arr[i:min(i + min_run, n)])

    size = min_run
    while size < n:
        for start in range(0, n, size * 2):
            mid = min(n, start + size)
            end = min(n, start + size * 2)
            merged_array = merge(arr[start:mid], arr[mid:end])
            arr[start:start + len(merged_array)] = merged_array
        size *= 2

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result
```

```python
result = {}
```

```python
data_1 = random.choices(range(1, 10), k=10)
data_2 = random.choices(range(1, 1000), k=1000)
data_3 = random.choices(range(1, 100000), k=10000)

data_4 = random.choices(range(1, 2), k=10)
data_5 = random.choices(range(1, 10), k=1000)
data_6 = random.choices(range(1, 100), k=10000)
```

```python
for data in [data_1, data_2, data_3]:
    data_sorted = data.copy()
    data_tim = data.copy()
    data_merge = data.copy()
    data_insertion = data.copy()

    sorted_algoritm = timeit.timeit(lambda: sorted(data_sorted), number=10)
    timsort_algoritm = timeit.timeit(lambda: timsort(data_tim), number=10)
    merge_sort_algoritm = timeit.timeit(lambda: merge_sort(data_merge), number=10)
    insertion_sort_algoritm = timeit.timeit(lambda: insertion_sort(data_insertion), number=10)

    result[len(data)] = {
        'sorted': sorted_algoritm,
        'timsort': timsort_algoritm,
        'merge_sort': merge_sort_algoritm,
        'insertion_sort': insertion_sort_algoritm
    }

for data in [data_4, data_5, data_6]:
    data_sorted = data.copy()
    data_tim = data.copy()
    data_merge = data.copy()
    data_insertion = data.copy()

    sorted_algoritm = timeit.timeit(lambda: sorted(data_sorted), number=10)
    timsort_algoritm = timeit.timeit(lambda: timsort(data_tim), number=10)
    merge_sort_algoritm = timeit.timeit(lambda: merge_sort(data_merge), number=10)
    insertion_sort_algoritm = timeit.timeit(lambda: insertion_sort(data_insertion), number=10)

    result[f'{len(data)}_dub'] = {
        'sorted': sorted_algoritm,
        'timsort': timsort_algoritm,
        'merge_sort': merge_sort_algoritm,
        'insertion_sort': insertion_sort_algoritm
    }
```
